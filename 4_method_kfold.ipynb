```python
"""
四种对数据集进行k折划分的方法：Kfold，GroupKFold，StratifiedKfold，Stratify GroupKFold
"""
```

```python
#Kfold,普通的k折交叉验证,是对数据集顺序按顺序K折划分
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
train = pd.read_csv("/home/xhzhu/data/siim/train.csv")
#用1000个来测试
# t_train = train[:1000]
t_train = train
```

```python
kfold_t_train = t_train.copy()
kfold_t_train.loc[:, 'fold'] = 0
kf = KFold(n_splits=5, shuffle=True, random_state=8)
for fold_number, (train_index, test_index) in enumerate(kf.split(t_train)):
    print("TRAIN:", train_index, "TEST:", test_index)
    kfold_t_train.loc[kfold_t_train.iloc[test_index].index, 'fold'] = fold_number
```

```python
kfold_t_train
```

```python
print(len(kfold_t_train[kfold_t_train['fold']==0]), "fold0")
print(len(kfold_t_train[kfold_t_train['fold']==1]), "fold1")
print(len(kfold_t_train[kfold_t_train['fold']==2]), "fold2")
print(len(kfold_t_train[kfold_t_train['fold']==3]), "fold3")
print(len(kfold_t_train[kfold_t_train['fold']==4]), "fold4")
```

```python
"""
tratifiedKFold，平衡label的k折交叉验证。在类别label不平衡时使用，如0、1分类时，0的样本量和1的样本量差异悬殊。
分层采样，确保训练集，验证集中各类别样本的比例与原始数据集中相同；
"""
from sklearn.model_selection import StratifiedKFold
StratifiedKFold_t_train = t_train.copy()
StratifiedKFold_t_train.loc[:, 'fold'] = 0
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=3)
for fold_number, (train_index, test_index) in enumerate(skf.split(t_train.iloc[:,:7],t_train.iloc[:,7])):
    print("TRAIN:", train_index, "TEST:", test_index)
    StratifiedKFold_t_train.loc[StratifiedKFold_t_train.iloc[test_index].index, 'fold'] = fold_number
```

```python
StratifiedKFold_t_train
```

```python
print(len(StratifiedKFold_t_train[(StratifiedKFold_t_train['fold']==0) & (StratifiedKFold_t_train['target'] == 1)]), "fold0")
print(len(StratifiedKFold_t_train[(StratifiedKFold_t_train['fold']==1) & (StratifiedKFold_t_train['target'] == 1)]), "fold1")
print(len(StratifiedKFold_t_train[(StratifiedKFold_t_train['fold']==2) & (StratifiedKFold_t_train['target'] == 1)]), "fold2")
print(len(StratifiedKFold_t_train[(StratifiedKFold_t_train['fold']==3) & (StratifiedKFold_t_train['target'] == 1)]), "fold3")
print(len(StratifiedKFold_t_train[(StratifiedKFold_t_train['fold']==4) & (StratifiedKFold_t_train['target'] == 1)]), "fold4")
```

```python
"""
GroupKFold ,比如df数据集中，有一列用户的编号user_id，而其中每个user_id可能有多个样本，
我们在使用k折交叉验证的时候希望对user_id进行拆分（类似groupby），将一个user_id的所有样本同时放在train中或者vaild中
"""
from sklearn.model_selection import GroupKFold
GroupKFold_t_train = t_train.copy()
GroupKFold_t_train.loc[:, 'fold'] = 0
group_kfold = GroupKFold(n_splits=5)
for fold_number, (train_index, test_index) in enumerate(group_kfold.split(t_train.iloc[:,:7],t_train.iloc[:,7],t_train['patient_id'])):
    print("TRAIN:", train_index, "TEST:", test_index)
    GroupKFold_t_train.loc[GroupKFold_t_train.iloc[test_index].index, 'fold'] = fold_number
```

```python
"""
Stratify GroupKFold,groupkfold和stratifykfold的合体
"""
import random
from tqdm import tqdm  
from collections import Counter, defaultdict

def stratified_group_k_fold(X, y, groups, k, seed=None):
    """ https://www.kaggle.com/jakubwasikowski/stratified-group-k-fold-cross-validation """
    labels_num = np.max(y) + 1
    y_counts_per_group = defaultdict(lambda: np.zeros(labels_num))
    y_distr = Counter()
    for label, g in zip(y, groups):
        y_counts_per_group[g][label] += 1
        y_distr[label] += 1

    y_counts_per_fold = defaultdict(lambda: np.zeros(labels_num))
    groups_per_fold = defaultdict(set)

    def eval_y_counts_per_fold(y_counts, fold):
        y_counts_per_fold[fold] += y_counts
        std_per_label = []
        for label in range(labels_num):
            label_std = np.std([y_counts_per_fold[i][label] / y_distr[label] for i in range(k)])
            std_per_label.append(label_std)
        y_counts_per_fold[fold] -= y_counts
        return np.mean(std_per_label)
    
    groups_and_y_counts = list(y_counts_per_group.items())
    random.Random(seed).shuffle(groups_and_y_counts)

    for g, y_counts in tqdm(sorted(groups_and_y_counts, key=lambda x: -np.std(x[1])), total=len(groups_and_y_counts)):
        best_fold = None
        min_eval = None
        for i in range(k):
            fold_eval = eval_y_counts_per_fold(y_counts, i)
            if min_eval is None or fold_eval < min_eval:
                min_eval = fold_eval
                best_fold = i
        y_counts_per_fold[best_fold] += y_counts
        groups_per_fold[best_fold].add(g)

    all_groups = set(groups)
    for i in range(k):
        train_groups = all_groups - groups_per_fold[i]
        test_groups = groups_per_fold[i]

        train_indices = [i for i, g in enumerate(groups) if g in train_groups]
        test_indices = [i for i, g in enumerate(groups) if g in test_groups]

        yield train_indices, test_indices
```

```python
df_folds = t_train.copy()
df_folds['patient_id'] = df_folds['patient_id'].fillna(df_folds['image_name'])
df_folds['sex'] = df_folds['sex'].fillna('unknown')
df_folds['anatom_site_general_challenge'] = df_folds['anatom_site_general_challenge'].fillna('unknown')
df_folds['age_approx'] = df_folds['age_approx'].fillna(round(df_folds['age_approx'].mean()))

patient_id_2_count = df_folds[['patient_id', 'image_name']].groupby('patient_id').count()['image_name'].to_dict()

df_folds = df_folds.set_index('image_name')

def get_stratify_group(row):
    stratify_group = row['sex']
#     stratify_group += f'_{row["anatom_site_general_challenge"]}'
  #  stratify_group += f'_{row["source"]}'
    stratify_group += f'_{row["target"]}'
    patient_id_count = patient_id_2_count[row["patient_id"]]
    if patient_id_count > 80:
        stratify_group += f'_80'
    elif patient_id_count > 60:
        stratify_group += f'_60'
    elif patient_id_count > 50:
        stratify_group += f'_50'
    elif patient_id_count > 30:
        stratify_group += f'_30'
    elif patient_id_count > 20:
        stratify_group += f'_20'
    elif patient_id_count > 10:
        stratify_group += f'_10'
    else:
        stratify_group += f'_0'
    return stratify_group

df_folds['stratify_group'] = df_folds.apply(get_stratify_group, axis=1)
#对数据进行编码转换，转换后需要查询标签和编码的对应关系
#可参考博客理解：https://blog.csdn.net/u010652755/article/details/105594204
df_folds['stratify_group'] = df_folds['stratify_group'].astype('category').cat.codes
print(df_folds['stratify_group'])
```

```python
df_folds.loc[:, 'fold'] = 0
skf = stratified_group_k_fold(X=df_folds.index, y=df_folds['stratify_group'], groups=df_folds['patient_id'], k=5, seed=42)
for fold_number, (train_index, test_index) in enumerate(skf):
    #print("TRAIN:", train_index, "TEST:", test_index)
    df_folds.loc[df_folds.iloc[test_index].index, 'fold'] = fold_number
```

```python
print(len(df_folds[(df_folds['fold']==0) & (df_folds['target'] == 1)]), "fold0")
print(len(df_folds[(df_folds['fold']==1) & (df_folds['target'] == 1)]), "fold1")
print(len(df_folds[(df_folds['fold']==2) & (df_folds['target'] == 1)]), "fold2")
print(len(df_folds[(df_folds['fold']==3) & (df_folds['target'] == 1)]), "fold3")
print(len(df_folds[(df_folds['fold']==4) & (df_folds['target'] == 1)]), "fold4")
```
